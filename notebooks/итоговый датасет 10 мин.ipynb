4. y<sub>4</sub> = % –ø—Ä–∏—Ä–æ—Å—Ç –∑–∞ 10 –º–∏–Ω—É—Ç

```python
# y_1 - shift –Ω–∞ 10 –º–∏–Ω—É—Ç –≤–ø–µ—Ä–µ–¥
result['avg_price'] = ((result['<OPEN>'] + result['<CLOSE>'] + result['<HIGH>'] + result['<LOW>'])/4)
result['target_1'] = result['avg_price'].shift(-1)
result['ticker shift'] = result['<TICKER>'].shift(-1)
result.loc[result['<TICKER>'] == result['ticker shift'], 'y_1'] = result['target_1']
result.loc[result['<TICKER>'] != result['ticker shift'], 'y_1'] = None
```

```python
# y_2 - shift —Ä–∞–∑–Ω–∏—Ü—ã –Ω–∞ 10 –º–∏–Ω—É—Ç –≤–ø–µ—Ä–µ–¥

result['target_2'] = result['y_1'] - result['avg_price']
result.loc[result['<TICKER>'] == result['ticker shift'], 'y_2'] = result['target_2']
result.loc[result['<TICKER>'] != result['ticker shift'], 'y_2'] = None
```

```python
# y_3 - —Ç—Ä–µ–Ω–¥ (-1 –∏–ª–∏ +1)

result.loc[result['y_2'] >= 0, 'y_3'] = 1
result.loc[result['y_2'] < 0, 'y_3'] = -1
```

```python
# y_4 - % –ø—Ä–∏—Ä–æ—Å—Ç–∞ –Ω–∞ 10 –º–∏–Ω—É—Ç –≤–ø–µ—Ä–µ–¥

result['y_4'] = result['y_1'] / result['avg_price'] - 1
```

```python
# y_5 - shift –Ω–∞ 20 –º–∏–Ω—É—Ç –≤–ø–µ—Ä–µ–¥

result['target_3'] = result['avg_price'].shift(-2)
result['ticker shift 2'] = result['<TICKER>'].shift(-2)
result.loc[result['<TICKER>'] == result['ticker shift 2'], 'y_5'] = result['target_3']
result.loc[result['<TICKER>'] != result['ticker shift 2'], 'y_5'] = None
```

```python
# y_6 - shift —Ä–∞–∑–Ω–∏—Ü—ã –Ω–∞ 20 –º–∏–Ω—É—Ç –≤–ø–µ—Ä–µ–¥

result['target_4'] = result['y_5'] - result['avg_price']
result.loc[result['<TICKER>'] == result['ticker shift 2'], 'y_6'] = result['target_4']
result.loc[result['<TICKER>'] != result['ticker shift 2'], 'y_6'] = None
```

```python
# y_7 - —Ç—Ä–µ–Ω–¥ (-1 –∏–ª–∏ +1) –∑–∞ 20 –º–∏–Ω—É—Ç

result.loc[result['y_6'] >= 0, 'y_7'] = 1
result.loc[result['y_6'] < 0, 'y_7'] = -1
```

```python
# y_8 - % –ø—Ä–∏—Ä–æ—Å—Ç–∞ –Ω–∞ 20 –º–∏–Ω—É—Ç –≤–ø–µ—Ä–µ–¥

result['y_8'] = result['y_5'] / result['avg_price'] - 1
```

```python
!pip install pymorphy2
import pymorphy2
import re
morph = pymorphy2.MorphAnalyzer()
def lemmatize(text):
    """
  –°–æ–∑–¥–∞—ë–º –º–µ—Ç–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä—É–µ—Ç —Ç–µ–∫—Å—Ç.
  –ê—Ä–≥—É–º–µ–Ω—Ç:
  text - —Ç–µ–∫—Å—Ç, –∫–æ—Ç–æ—Ä—ã–π –Ω—É–∂–Ω–æ –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞—Ç—å
  –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
  res - –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç
    """
    try:
        words = re.split(r'\W+', text) # —Ä–∞–∑–±–∏–≤–∞–µ–º —Ç–µ–∫—Å—Ç –Ω–∞ —Å–ª–æ–≤–∞
    except:
        return text
    res = str()
    for word in words:
        p = morph.parse(word)[0]
        res += (p.normal_form) + ' '
    return res
```

```python
result['lem_message'] = result['raw_news'].apply(lambda x: lemmatize(x))
```
