```python
def about_companies(x):
    lable = []
    i1, i2, i3, i4, i5, i6 = 0, 0, 0, 0, 0, 0
    for i in x:
        k=0
        lable1 = []
        if '–ª—É–∫–æ–π–ª' in str(i) or 'lkoh' in str(i):
            i5+=1
            k+=1
            lable1.append('lkoh')
        if '–≤—Ç–±' in str(i) or 'vtbr' in str(i):
            i1+=1
            k+=1
            lable1.append('vtbr')
        if '—Å–±–µ—Ä' in str(i) or 'sber' in str(i):
            i2+=1
            k+=1
            lable1.append('sber')
        if '–≥–∞–∑–ø—Ä–æ–º' in str(i) or 'gazp' in str(i):
            i3 +=1
            k+=1
            lable1.append('gazp')
        if '—è–Ω–¥–µ–∫—Å' in str(i) or 'yndx' in str(i):
            i4 +=1
            k+=1
            lable1.append('yndx')
        if '—Ä–æ—Å–Ω–µ—Ñ—Ç' in str(i) or 'rosn' in str(i):
            i6 +=1
            k+=1
            lable1.append('rosn')
        if k ==0:
            lable1.append('none')
        lable.append(lable1)
            

    #for i, j in zip(['–ª—É–∫–æ–π–ª', '–≤—Ç–±', '—Å–±–µ—Ä', '–≥–∞–∑–ø—Ä–æ–º', '—è–Ω–¥–µ–∫—Å', '—Ä–æ—Å–Ω–µ—Ñ—Ç—å'], [i5, i1, i2, i3, i4, i6]):
        #print(i + ': ' + str(j))
    return lable
```

```python
def to_str(x):
    return str(x)[1:-1:]
```

```python
x = list(df['message'])
about_companies(x)
```

```python
df['comp'] = about_companies(x)
```

```python
shares = pd.read_csv('YNDX_180101_211223.csv', sep = ';')
```

```python
df['time'] = df['date'].apply(lambda x: str(x)[-8:])
```

```python
df['<TIME>'] = df['time'].apply(lambda x: str(x)[:2]+':00:00')
```

```python
df['<DATE>'] = df['date'].apply(lambda x: str(x)[:10].replace('-', ''))
```

```python
df_new = df[['date', 'message', 'comp', '<DATE>', '<TIME>']]
```

```python
df_new['period'] = df['<DATE>']+' '+df['<TIME>']
```

```python
df_new
```

```python
shares['period'] = shares['<DATE>'].astype(str) +' '+shares['<TIME>'].astype(str)
```

```python
shares['delta'] = shares['<CLOSE>'] - shares['<OPEN>']
```

```python
shares['abs_delta'] = abs(shares['<CLOSE>'] - shares['<OPEN>'])
```

```python
def yandex(x):
    if 'yndx' in x:
        return 1
    else:
        return 0
```

```python
df_new['about_yndx'] = df_new['comp'].apply(lambda x: yandex(x))
```

```python
df_yandex = shares.merge(df_new[['about_yndx', 'period']].groupby("period", as_index=False).sum())
```

```python
df_yandex['about_yndx'].value_counts()
```

```python
df_yandex['news_ind'] = df_yandex['about_yndx'].apply(lambda x: 1 if x>0 else 0)
```
